```python
import torch
import torch.nn as nn
```

```python
class VanillaCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        
        
        self.features = nn.Sequential(
            nn.Conv2d(3,32,3, padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(32,64,3, padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        
        self.classifier = nn.Sequential(
            nn.Linear(64*8*8,128),
            nn.ReLU(),
            nn.Linear(128,num_classes)
        )
    
    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x,1)
        return self.classifier(x)
    
    
```

`nn.Conv2d` has some parameters</br>
`in_channels = no of input feature maps like RGB has 3 grayscale has 1`</br>
`out_channels = no of filters = no of features maps output`</br>
`kernel_size = size of sliding window`</br>
`stride = step size`</br>
`padding = adds border around input`</br>


Order of the parameters are `(in_channels, out_channels, kernel_size, stride, padding=...)`</br>
